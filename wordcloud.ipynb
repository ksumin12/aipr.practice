```python
with open('/content/drive/My Drive/Colab Notebooks/data/news.txt', 'r', encoding='utf-8') as f:
    #news.txt 파일을 읽기모드로 열어서 텍스트 데이터를 읽어와 text 변수에 저장
    text = f.read()
```

```python
okt = Okt()
nouns = okt.nouns(text) #Okt 클래스의 nouns 메서드를 호출해 입력된 text에서 명사를 추출
```

```python
stopwords = ['것', '안' , '렘수', '중', '수', '이', '또', '움직','을', '은', '.', ',' ]

words = text.split() #text 변수에 있는 텍스트를 split 메서드 사용해 공백 기준으로 분리, 각각의 단어 또는 토큰으로 이루어진 목록 생성

filtered_words = [word for word in words if word not in stopwords]
#filtered_words 변수에 words 목록에서 stopwords에 포함된 단어를 제외, 남은 단어들로 이루어진 목록이 저장됨
#words 목록 순회, stopwords에 포함되지 않은 단어들만을 filtered_words 목록으로 추출
```

```python
c = Counter(filtered_words)
#words 리스트 내 단어 빈도 계산하여 각 단어가 몇번 나왔는지 세어주는 Counter 객체 생성
```

```python

wc = WordCloud(font_path='/content/drive/My Drive/Colab Notebooks/data/KBO_bold.ttf', width=400, height=400, scale=2.0, max_font_size=250)
#word cloud 객체를 생성하고 malgun 폰트를 사용해 크기 스케일 최대폰트 크기 설정

gen = wc.generate_from_frequencies(c)
#Counter의 객체인 c를 기반으로 워드클라우드 생성

plt.figure()
plt.imshow(gen)
```
