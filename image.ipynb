```python
from PIL import Image
from PIL import ImageFilter
```

```python
img = Image.open("C:/baseball.jpg")

img.show() #이미지 불러오기 
```

```python
width = 400
height = 256
# 원하는 새로운 크기 설정

resized_img = img.resize((width, height))  

# 리샘플링된 이미지 저장
#resized_img.save('resampled_image.jpg')

resized_img.show()
```

```python
img_resize = img.resize((int(img.width / 2), int(img.height / 2))) #원본사진 반으로 줄임, 정수값으로 설정->int사용

img_resize.show()
```

```python
#이미지 자르기

import cv2

image_copy = img.copy() #원본파일복사
```

```python
image_cut = img.crop((100,600,200,700)) #원본사진 자르기
image_cut.show()
```

```python
#이미지 회전

image_copy.rotate(90)
```

```python
#이미지 좌우 대칭

FlipImage = image_copy.transpose(Image.FLIP_LEFT_RIGHT)

FlipImage.show()
```

```python
#이미지 상하 대칭

FlipImage2 = image_copy.transpose(Image.FLIP_TOP_BOTTOM)

FlipImage2.show()
```

```python
#이미지 필터링

#blur

blur = image_copy.filter(ImageFilter.BLUR) #인자값 없음

blur.show()

#GaussianBlur
blur3 = image_copy.filter(ImageFilter.GaussianBlur(5))
 
blur3.show()
```

```python
#BoxBlur / 흐릿한 값의 정도 지정 가능
blur2 = image_copy.filter(ImageFilter.BoxBlur(7)) #인자값 넣어야함  

blur2.show()
```

```python
#GaussianBlur
blur3 = image_copy.filter(ImageFilter.GaussianBlur(5))
 
blur3.show()
```

```python
#image contour

import cv2
import numpy as np

# 이미지 불러오기
img = cv2.imread('C:/mouww.jpg', cv2.IMREAD_UNCHANGED)

# 이미지를 회색조로 변환
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 임계값 설정
thresh = 100

# 임계값 이미지 가져오기
ret, thresh_img = cv2.threshold(img_grey, thresh, 255, cv2.THRESH_BINARY)

# 윤곽 찾기
contours, hierarchy = cv2.findContours(thresh_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# 빈 이미지 생성
img_contours = np.zeros(img.shape, dtype=np.uint8)

# 윤곽을 빈 이미지에 그리기
cv2.drawContours(img_contours, contours, -1, (0, 255, 0), 3)

# 이미지 저장
cv2.imwrite('contours_image.jpg', img_contours)

# 이미지 출력
cv2.imshow('Contours Image', img_contours)
cv2.waitKey(0)  # 키보드 입력을 대기하며, 아무 키나 누를 때까지 대기

# 창 닫기
cv2.destroyAllWindows()
```

```python
#넘파이배열 이미지 변환
import io

image = Image.open("C:/Users/82105/Downloads/cat.jpg")

# 이미지를 바이트 배열로 변환
image_byte_array = io.BytesIO()
image.save(image_byte_array, format="JPEG")  # 이미지 포맷은 필요에 따라 변경

# 바이트 배열을 얻기
image_bytes = image_byte_array.getvalue()

print(image_bytes)
```

```python
from PIL import Image
import numpy as np

# 넘파이 배열 생성 (예: 흑백 이미지)
# 이미지 크기: 100x100
# 데이터 범위: 0 (검정) ~ 255 (흰색)
image_data = np.random.randint(0, 256, (500,500), dtype=np.uint8)

# 넘파이 배열을 이미지로 변환
image33 = Image.fromarray(image_data)

# 이미지 저장 또는 표시

image33.show()  # 이미지를 뷰어로 표시
```
